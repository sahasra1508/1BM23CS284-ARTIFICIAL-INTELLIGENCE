```python
board = {1:' ', 2:' ', 3:' ',
         4:' ', 5:' ', 6:' ',
         7:' ', 8:' ', 9:' '}

def printBoard(board):
    print(board[1]+'|'+board[2]+'|'+board[3])
    print('-+-+-')
    print(board[4]+'|'+board[5]+'|'+board[6])
    print('-+-+-')
    print(board[7]+'|'+board[8]+'|'+board[9])
    print('\n')

def spaceFree(pos):
    return board[pos] == ' '

def checkWin():
    # Check rows, columns and diagonals for win
    win_conditions = [
        (1,2,3), (4,5,6), (7,8,9),
        (1,4,7), (2,5,8), (3,6,9),
        (1,5,9), (3,5,7)
    ]
    for a,b,c in win_conditions:
        if board[a] == board[b] == board[c] != ' ':
            return True
    return False

def checkMoveForWin(move):
    win_conditions = [
        (1,2,3), (4,5,6), (7,8,9),
        (1,4,7), (2,5,8), (3,6,9),
        (1,5,9), (3,5,7)
    ]
    for a,b,c in win_conditions:
        if board[a] == board[b] == board[c] == move:
            return True
    return False

def checkDraw():
    for pos in board:
        if board[pos] == ' ':
            return False
    return True

def insertLetter(letter, position):
    if spaceFree(position):
        board[position] = letter
        printBoard(board)
    else:
        print('Position taken, please pick a different position.')
        position = int(input('Enter new position: '))
        insertLetter(letter, position)

player = 'O'
bot = 'X'

def playerMove():
    position = int(input('Enter position for O (1-9): '))
    insertLetter(player, position)

def compMove():
    bestScore = -1000
    bestMove = 0
    for key in board.keys():
        if board[key] == ' ':
            board[key] = bot
            score = minimax(board, False)
            board[key] = ' '
            if score > bestScore:
                bestScore = score
                bestMove = key
    insertLetter(bot, bestMove)

def minimax(board, isMaximizing):
    if checkMoveForWin(bot):
        return 1
    elif checkMoveForWin(player):
        return -1
    elif checkDraw():
        return 0

    if isMaximizing:
        bestScore = -1000
        for key in board.keys():
            if board[key] == ' ':
                board[key] = bot
                score = minimax(board, False)
                board[key] = ' '
                if score > bestScore:
                    bestScore = score
        return bestScore
    else:
        bestScore = 1000
        for key in board.keys():
            if board[key] == ' ':
                board[key] = player
                score = minimax(board, True)
                board[key] = ' '
                if score < bestScore:
                    bestScore = score
        return bestScore

# Main game loop
while True:
    compMove()
    if checkWin():
        print("Bot wins!")
        break
    if checkDraw():
        print("It's a draw!")
        break

    playerMove()
    if checkWin():
        print("You win!")
        break
    if checkDraw():
        print("It's a draw!")
        break
```

```python
def vacuum_world():
    # Initializing goal_state: 0 means clean, 1 means dirty
    goal_state = {'A': '0', 'B': '0'}
    cost = 0

    # User input: location and status
    location_input = input("Enter Location of Vacuum (A or B): ")  # Vacuum's current location
    status_input = input("Enter status of " + location_input + " (1 for Dirty, 0 for Clean): ")  # Status of the current location
    status_input_complement = input("Enter status of other room (1 for Dirty, 0 for Clean): ")  # Status of the other room

    print("Initial Location Condition: " + str(goal_state))

    if location_input == 'A':
        print("Vacuum is placed in Location A")

        if status_input == '1':  # If Location A is dirty
            print("Location A is Dirty.")
            goal_state['A'] = '0'  # Clean A
            cost += 1  # Cost for cleaning
            print("Cost for CLEANING A: " + str(cost))
            print("Location A has been Cleaned.")

            if status_input_complement == '1':  # If Location B is dirty
                print("Location B is Dirty.")
                print("Moving right to Location B.")
                cost += 1  # Cost for moving to Location B
                print("Cost for moving RIGHT: " + str(cost))
                goal_state['B'] = '0'  # Clean B
                cost += 1  # Cost for cleaning B
                print("Cost for SUCK: " + str(cost))
                print("Location B has been Cleaned.")
            else:
                print("Location B is already clean. No action needed.")

        elif status_input == '0':  # If Location A is already clean
            print("Location A is already clean.")

            if status_input_complement == '1':  # If Location B is dirty
                print("Location B is Dirty.")
                print("Moving right to Location B.")
                cost += 1  # Cost for moving to Location B
                print("Cost for moving RIGHT: " + str(cost))
                goal_state['B'] = '0'  # Clean B
                cost += 1  # Cost for cleaning B
                print("Cost for SUCK: " + str(cost))
                print("Location B has been Cleaned.")
            else:
                print("Location B is already clean. No action needed.")

    elif location_input == 'B':
        print("Vacuum is placed in Location B")

        if status_input == '1':  # If Location B is dirty
            print("Location B is Dirty.")
            goal_state['B'] = '0'  # Clean B
            cost += 1  # Cost for cleaning B
            print("Cost for CLEANING B: " + str(cost))
            print("Location B has been Cleaned.")

            if status_input_complement == '1':  # If Location A is dirty
                print("Location A is Dirty.")
                print("Moving left to Location A.")
                cost += 1  # Cost for moving to Location A
                print("Cost for moving LEFT: " + str(cost))
                goal_state['A'] = '0'  # Clean A
                cost += 1  # Cost for cleaning A
                print("Cost for SUCK: " + str(cost))
                print("Location A has been Cleaned.")
            else:
                print("Location A is already clean. No action needed.")

        elif status_input == '0':  # If Location B is already clean
            print("Location B is already clean.")

            if status_input_complement == '1':  # If Location A is dirty
                print("Location A is Dirty.")
                print("Moving left to Location A.")
                cost += 1  # Cost for moving to Location A
                print("Cost for moving LEFT: " + str(cost))
                goal_state['A'] = '0'  # Clean A
                cost += 1  # Cost for cleaning A
                print("Cost for SUCK: " + str(cost))
                print("Location A has been Cleaned.")
            else:
                print("Location A is already clean. No action needed.")

    else:
        print("Invalid input. Please enter either 'A' or 'B'.")

    # Print the goal state and total cost
    print("\nGOAL STATE: ")
    print(goal_state)
    print("Performance Measurement: " + str(cost))

# Run the vacuum world simulation
vacuum_world()
```
